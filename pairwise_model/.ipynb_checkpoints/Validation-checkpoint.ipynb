# Validation

The first validation is to see how the test error behave while training the model. Runtime for 3000 iterations 4300 seconds

```python
%matplotlib inline
from load_data import *
from fit_pairwise import *
from data_preperation import *
import matplotlib.pyplot as plt

#No validation set for experiemtnal data
#Currently have to manly set paths to correct data-set, default 3D simplex data.
train_s,train_n, test_s,test_n  = load_bpp_data(nr_seed=30,factor_test=0.1,split=True)

data = train_s.T
test = test_s.T


data = data
test = test

[m, n] = np.shape(data)
print(m,n)
J0 = np.zeros((n, n))
# J0 = np.random.normal(loc=0,scale=0.1,size=(n,n))


learning_rate = 1
# NB Chose M-samples such that it can be divded equally beteen all processes.
M_samples = m
print(m)
# Number of cores for multiprocess.
n_jobs = 2
entropy = fit_pairwise(data, J0, learning_rate, M_samples, n_jobs=n_jobs, test=test, save_loss=True)
# entropy = fit_pairwise(train,J0,learning_rate,M_samples,n_jobs=n_jobs,test=test,save_loss=True)
iter = 1
# Recommended to have twice as many as neuronss.
gibbs_steps = 2*n
# gibbs_steps = 8

verbose = str(gibbs_steps)+"_"+str(iter)+"_"+str(M_samples)+"_"+str(learning_rate)

start_time = time.time()

J, emp_cov , mod_cov, samples, train_error, test_error = entropy.fit_pairwise(iter, gibbs_steps)
print("--- %s seconds ---" % (time.time() - start_time))
```

## Learning curve

```python
I = range(0,iter+2)
fig = plt.figure()
ax = fig.add_subplot(111)
if test_error is not None:
    line2, = plt.plot(I, test_error, 'b--', label="Test error")
line1, = plt.plot(I, train_error, 'r--', label="Training error")
plt.ylabel("Max(error)")
plt.xlabel("Iteration")
first_legend = plt.legend(handles=[line1], loc=1)
ax = plt.gca().add_artist(first_legend)
if test_error is not None:
    plt.legend(handles=[line2], loc=4)
plt.title('Learning curve')
plt.savefig('figures/learning_curve/learning_curve.png')
plt.show()
```

Let's see how the likelihood between the training- and testa data differs.

```python
from likelihood_pairwise import likelihood_pairwise
#train_s,train_n, test_s,test_n  = load_bpp_data(nr_seed=30,factor_test=0.1,split=True)

#data = train_s.T
#test = test_s.T
#data = np.genfromtxt("train.csv", delimiter=",").T
#test = np.genfromtxt("test.csv", delimiter=",").T

#data = data[:,:10]
#test = test[:,:10]

[m, n] = np.shape(data)

J0 = np.zeros((n, n))



iter = 200



betas = np.linspace(0, 1, num=20)
log_like_ratios = []

for i in range(0,10):
    train_s,train_n, test_s,test_n  = load_bpp_data(nr_seed=30,factor_test=0.1,split=True,seed=i)
    data = train_s.T
    test = test_s.T
    #data = data[:,:10]
    #test = test[:,:10]
    [m, n] = np.shape(data)
    J0 = np.zeros((n, n))
    entropy = fit_pairwise(data, J0, learning_rate, M_samples, n_jobs=n_jobs, save_loss=True)
    J, emp_cov , mod_cov, samples, train_error = entropy.fit_pairwise(iter, gibbs_steps)
    neg_log_L_train = likelihood_pairwise(data, M_samples, J, betas)
    neg_log_L_test = likelihood_pairwise(test, M_samples, J, betas)
    log_like_ratios.append(neg_log_L_train/neg_log_L_test)

```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(log_like_ratios, 0)
ax.set_title("Likelihood ratios")
ax.set_ylabel("$L_{Train}/L_{Test}$")
plt.savefig('figures/box_plot/ed_'+str(verbose)+'.png')
plt.show()
```
